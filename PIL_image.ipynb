```python
#RED CHANNEL VALUES 0 to 255
plt.imshow(pic_red[:,:,0],cmap='gray')  # 0 means no red ,255 means full read
```

```python
#GREEN CHANNEL VALUES 0 to 255
plt.imshow(pic_red[:,:,1],cmap='gray')
```

```python
#BLUE CHANNEL VALUES 0 to 255
plt.imshow(pic_red[:,:,2],cmap='gray')
```
