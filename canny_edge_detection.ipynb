```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('dog_face.jpg')
```

```python
plt.imshow(img)
```

```python
edges = cv2.Canny(image=img,threshold1=127,threshold2=127)
```

```python
plt.imshow(edges)
```

```python
edges = cv2.Canny(image=img,threshold1=0,threshold2=255)
```

```python
plt.imshow(edges)
```

```python
med_val = np.median(img)
```

```python
med_val
```

```python
lower = int(max(0,0.7*med_val))
upper = int(min(255,1.3*med_val))
```

```python
edges = cv2.Canny(image=img,threshold1=lower+100,threshold2=upper+100)
```

```python
plt.imshow(edges)
```

```python
upper
```

```python
lower
```

```python
blurred_img = cv2.blur(img,ksize=(5,5))
```

```python
edges = cv2.Canny(image=blurred_img,threshold1=lower,threshold2=upper+50)
```

```python
plt.imshow(edges)
```
