```python
dst = cv2.cornerHarris(src=real_chess,blockSize=2,ksize=3,k=0.04)
```

```python
dst = cv2.dilate(dst,None)
```

```python
flat_chess[dst>0.01*dst.max()] = [255,0,0]
```
