```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
flat_chess = cv2.imread('chess5.jpg')
plt.imshow(flat_chess)
```

```python
found,corners = cv2.findChessboardCorners(flat_chess,(7,7))
```

```python
found
```

```python
cv2.drawChessboardCorners(flat_chess,(7,7),corners,found)
```

```python
plt.imshow(flat_chess)
```
