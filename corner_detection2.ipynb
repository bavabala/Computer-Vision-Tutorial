```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
real_chess  = cv2.imread('real_chess.jpg')
real_chess = cv2.cvtColor(real_chess,cv2.COLOR_BGR2RGB)
```

```python
flat_chess = cv2.imread('chess5.jpg')
flat_chess = cv2.cvtColor(flat_chess,cv2.COLOR_BGR2RGB)
```

```python
gray_flat_chess = cv2.cvtColor(flat_chess,cv2.COLOR_BGR2GRAY)
gray_flat_chess = cv2.cvtColor(real_chess,cv2.COLOR_BGR2GRAY)
```

```python
flat_chess.min()
```

```python
corners = cv2.goodFeaturesToTrack(gray_flat_chess,64,0.01,600)
```

```python
corners = np.int0(corners)
```

```python
for i in corners:
    x,y = i.ravel()
    cv2.circle(flat_chess,(x,y),3,(255,0,0),-1)
```

```python
plt.imshow(flat_chess)
```

```python
corners = cv2.goodFeaturesToTrack(gray_flat_chess,90,0.01,10)
```

```python
corners = np.int0(corners)
```

```python
for i in corners:
    x,y = i.ravel()
    cv2.circle(flat_chess,(x,y),3,(255,0,0),-1)
```

```python
plt.imshow(flat_chess)
```
